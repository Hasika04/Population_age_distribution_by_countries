# Task 01 â€“ Age Group Distribution by Country

In this task, we visualize the population distribution by age groups across the top 10 most populous countries using a stacked bar chart.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/LENOVO/Downloads/population_by_age_group.csv")
df.head()
```

```python
df['Country'].nunique()
```

The dataset contains 252 countries. To keep the chart readable and focused, we visualized only the **top 10 countries by total population**.

```python
top_10_df = df.sort_values(by= 'Total', ascending= False).head(10)
```

```python
# Set figure size
plt.figure(figsize=(12, 6))

# Plot stacked bar
age_groups = ['65+', '25-64 years', '15-24 years', '5-14 years', '0-4 years']
top_10_df_plot = top_10_df.set_index('Country')[age_groups]

top_10_df_plot.plot(kind='bar', stacked=True, figsize=(14, 6), colormap='tab20c')
plt.title("Population by Age Group in Selected Countries")
plt.ylabel("Population")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title="Age Group")
plt.show()
```

# Insights

- The 25â€“64 years group forms the largest chunk of population in most countries.
- Countries like Afghanistan have higher percentages in 0â€“14 years age group, indicating a younger population.
- Aging trends are more visible in countries with larger 65+ groups.
